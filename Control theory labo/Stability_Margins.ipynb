# Stability Margins

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from matplotlib import colors as mcolors

import package_LAB
from importlib import reload
package_LAB = reload(package_LAB)
from package_LAB import Process, Bode_Ls, Stability_Margins, IMC_TUNING
```

### General Data

```python
Kp_p = 0.5438749526909622
T1_p = 184.2423025000344
T2_p = 6.478747375951926e-5
theta_p = 2.9999487500471216
alpha = 0.1

omega = np.logspace(-5, 2, 70000)
```

## Gamma = 0.2

### L(s) Bode Plot

```python
gamma = 0.2
Kc, Ti, Td = IMC_TUNING(gamma, Kp_p, T1_p, T2_p, theta_p, method='SOPD-PID')

print(Kc)
print(Ti)
print(Td)

P1 = Process({}) 

P1.parameters['Kp'] = Kp_p
P1.parameters['T1'] = T1_p
P1.parameters['T2'] = T2_p
P1.parameters['theta'] = theta_p
P1.parameters['Kd'] = Td
P1.parameters['Kp_c'] = Kc
P1.parameters['Ki'] = Ti

Bode_Ls(P1,omega)
```

### L(s) Stability Margins

```python
pm, gm =Stability_Margins(P1,omega)
print("The gain stability margin is " +str(pm)+ " dB")
print("The phase stability margin is " +str(gm)+ "")
```
